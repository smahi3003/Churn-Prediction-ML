## Customer churn prediction
Customer churn prediction identifies the chances of customers dropping the company's products and services. Most businesses with repeat customers or customers in subscriptions want to retain the customers. Thus, tracking the customers who drop the subscription plan and those who proceed with the service is necessary. For such an approach, the organization should understand the client's behavior and the features that result in the client leaving. I will describe the steps involved in designing and implementing the churn prediction machine-learning model.

## Why churn prediction is crucial
It is necessary for any organization handling repeat customers to come up with ways of keeping the existing clients. The strategy is important as customer churn is costly, and it is costlier acquiring a new client than holding on to an existing client. Take the example of an internet service (ISP) provider who has gained a new user. They will need equipment and technicians for linking the new client to their network. The client will only pay the monthly subscription for continuing with their plan. In the event the user does not renew the service, the company will likely incur a loss, provided the trend repeats for multiple customers. Such an institution will likely record low monthly recurring revenue (MRR); hence, it will not be able to support the business. Therefore, an accurate churn prediction model should enable organizations to remain in business as they expand and gain customers.

### Visualisations

```python
def plot_categorical_distributions_plotly_grid(df, max_unique=20, max_plots=20):
    """
    Plots interactive bar charts of categorical-like columns in a 2-column grid layout.

    Parameters:
    - df: pandas DataFrame
    - max_unique: max number of unique values to be treated as categorical
    - max_plots: max number of plots to show
    """
    # Identify categorical-like columns
    cat_like_cols = [
        col for col in df.columns
        if df[col].dtype == 'object' or df[col].nunique() <= max_unique
    ]

    # Limit the number of plots
    num_plots = min(len(cat_like_cols), max_plots)
    num_rows = (num_plots + 1) // 2  # 2 plots per row
    vertical_spacing = min(0.04, 1 / (num_rows + 2))  # prevent layout errors

    # Create subplot grid
    fig = make_subplots(
        rows=num_rows,
        cols=2,
        subplot_titles=cat_like_cols[:num_plots],
        vertical_spacing=vertical_spacing,
        horizontal_spacing=0.1
    )

    # Add bar plots to the grid
    for i, col in enumerate(cat_like_cols[:num_plots]):
        row = (i // 2) + 1
        col_num = (i % 2) + 1
        counts = df[col].value_counts().reset_index()
        counts.columns = [col, 'count']

        fig.add_trace(
            go.Bar(x=counts[col], y=counts['count'], name=col, showlegend=False),
            row=row, col=col_num
        )

    # Layout adjustments
    fig.update_layout(
        height=350 * num_rows,
        title_text="Categorical Distributions",
        showlegend=False
    )

    fig.write_html(
        "categorical_distributions.html",
        include_plotlyjs="cdn"
    )
    fig.update_xaxes(tickangle=-45)
    fig.show()
```

```python
plot_categorical_distributions_plotly_grid(df)
```

## Churn distribution for categorical values

```python
def plot_churn_distribution(df, target_col='Churn', max_unique=20, max_plots=20):
    """
    Plots churn (binary target) distribution across categorical-like columns.

    Parameters:
    - df: pandas DataFrame
    - target_col: column name of the binary target (e.g., 'Churn')
    - max_unique: max unique values to be treated as categorical
    - max_plots: max number of categorical columns to plot
    """
    cat_like_cols = [
        col for col in df.columns
        if col != target_col and (df[col].dtype == 'object' or df[col].nunique() <= max_unique)
    ]

    num_plots = min(len(cat_like_cols), max_plots)
    num_rows = (num_plots + 1) // 2
    vertical_spacing = min(0.04, 1 / (num_rows + 2))

    fig = make_subplots(
        rows=num_rows, cols=2,
        subplot_titles=cat_like_cols[:num_plots],
        vertical_spacing=vertical_spacing,
        horizontal_spacing=0.1
    )

    for i, col in enumerate(cat_like_cols[:num_plots]):
        row = (i // 2) + 1
        col_num = (i % 2) + 1

        churn_dist = (
            df.groupby([col, target_col])
            .size()
            .reset_index(name='count')
        )

        total_counts = churn_dist.groupby(col)['count'].transform('sum')
        churn_dist['percentage'] = churn_dist['count'] / total_counts * 100

        for churn_value in sorted(df[target_col].dropna().unique()):
            sub_data = churn_dist[churn_dist[target_col] == churn_value]
            fig.add_trace(
                go.Bar(
                    x=sub_data[col],
                    y=sub_data['percentage'],
                    name=f'{target_col}={churn_value}'
                ),
                row=row, col=col_num
            )

    fig.update_layout(
        height=400 * num_rows,
        title_text=f"Churn Distribution across Categorical Features in percentage",
        barmode='group'
    )
    fig.write_html(
        "categorical_distributions_with_churn.html",
        include_plotlyjs="cdn"
    )
    fig.update_xaxes(tickangle=-45)
    fig.show()
```

```python
plot_churn_distribution(df, target_col='churn')
```
